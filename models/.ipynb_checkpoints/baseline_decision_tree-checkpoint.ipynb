```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import os
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd().parent
os.chdir(PROJECT_ROOT)
```

```python
X_train = pd.read_csv('data/processed/X_train.csv')
X_test = pd.read_csv('data/processed/X_test.csv')
y_train = pd.read_csv('data/processed/y_train.csv')
y_test = pd.read_csv('data/processed/y_test.csv')
```

```python
X_train_dt = X_train.copy()
X_test_dt  = X_test.copy()
```

```python
X_train_dt['Age'] = X_train_dt['Age'].fillna(X_train_dt['Age'].median())
X_test_dt['Age'] = X_test_dt['Age'].fillna(X_test_dt['Age'].median())
```

```python
X_train_dt['Embarked'] = X_train_dt['Embarked'] .fillna(
    X_train_dt["Embarked"].mode()[0]
)
```

```python
X_test_dt['Embarked'] = X_test_dt['Embarked'] .fillna(
    X_test_dt["Embarked"].mode()[0]
)
```

```python
X_train_dt = pd.get_dummies(
    X_train_dt,columns =['Sex','Embarked'],drop_first =True
)
X_test_dt = pd.get_dummies(
    X_test_dt,columns =['Sex','Embarked'],drop_first =True
)
```

```python
X_train_dt, X_test_dt = X_train_dt.align(X_test_dt, join="left", axis=1, fill_value=0)
```

```python
model = DecisionTreeClassifier(
    random_state = 42,
    max_depth = 5
)
model.fit(X_train,y_train)
```
